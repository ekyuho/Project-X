```python
import numpy as np

import os, sys
from IPython.display import display
from IPython.display import Image as _Imgdis
from PIL import Image
import pathlib
import cv2

import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras

image_width = 28
image_height = 28

digit_dir=pathlib.Path('../04_build_training_dataset/digits_image/')
model = tf.keras.models.load_model('model')
```

```python
def verify(fpath):
    answer = int(fpath.name[0])
    img = Image.open(fpath)
    img = img.resize((image_width, image_height))
    img = (np.expand_dims(img,0))/255.

    prediction = model.predict(img)
    if not answer == np.argmax(prediction[0]): 
        plt.figure()
        plt.imshow(img[0])
        plt.colorbar()
        plt.grid(False)
        plt.show()
        
        print('**** Oops. answer=', answer, 'predict=', np.argmax(prediction[0]), "file=", fpath)   
```

```python
paths = digit_dir.glob('*.png')
i=0
for fpath in paths:
    if i%500 == 0: print(' %d'%i)
    verify(fpath)
    i += 1
print("\nprocessed %d files."%i )
```
